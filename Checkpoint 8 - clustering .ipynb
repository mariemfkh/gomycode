```python
import pandas as pd
dataset=pd.read_csv(r"C:\Users\Ameur\Downloads\CC GENERAL.csv")
dataset
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isna().sum()
```

```python
dataset['CREDIT_LIMIT'].fillna(dataset['CREDIT_LIMIT'].mean(),inplace=True)
dataset
```

```python
dataset['MINIMUM_PAYMENTS'].fillna(dataset['MINIMUM_PAYMENTS'].mean(),inplace=True)
dataset 
```

```python
dataset.isna().sum()
```

```python
dataset.drop(["CUST_ID"],axis=1,inplace=True)
dataset
```

```python
from sklearn.cluster import AgglomerativeClustering 
model=AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(dataset)

```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
import matplotlib.pyplot as plt
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (dataset ['PURCHASES'] , dataset ["PAYMENTS"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("PURCHASES")
ax.set_ylabel("PAYMENTS")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("CREDIT CARD")
dend=shc.dendrogram(shc.linkage(dataset, method="complete"))
```

```python
from sklearn.cluster import KMeans
k= range(1,12)
Sum_of_squared_distances=[]
for i in k:
    kmean= KMeans(i)
    kmean.fit(dataset)
    Sum_of_squared_distances.append(kmean.inertia_)
plt.plot(k, Sum_of_squared_distances,)
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3, random_state=0) 
kmeans.fit(dataset)

labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(dataset)
print(kmeans.cluster_centers_)
```

```python
label=kmeans.predict(dataset)
print(kmeans.cluster_centers_)
```

```python
plt.scatter(dataset["PURCHASES"][label == 0],          
            dataset["PAYMENTS"][label == 0],s=10,c='magenta',label='Careful')
plt.scatter(dataset["PURCHASES"][label == 1],
           dataset["PAYMENTS"][label == 1],s=10,c='yellow',label='Standard')
plt.scatter(dataset["PURCHASES"][label == 2],
           dataset["PAYMENTS"][label == 2],s=10,c='green',label='Target')


plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=30, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('PURCHASES')
plt.ylabel('PAYMENTS')
plt.legend()
plt.show()
```
